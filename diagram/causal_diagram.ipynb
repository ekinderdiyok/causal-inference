```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from(['Member', 'CLV', 'Income'])

# Add edges (directed)
G.add_edges_from([('Member', 'CLV'), ('Income', 'CLV'), ('Income', 'Member')])

# Check if the graph is a DAG
is_dag = nx.is_directed_acyclic_graph(G)
print(f"The graph is a DAG: {is_dag}")


# Manually set positions for each node
pos = {
    'Member': (0, 0),
    'CLV': (0.2, 0),
    'Income': (0.1, 0.2)
}

fig, ax = plt.subplots(figsize=(8, 6))  # Create a new figure and axes
nx.draw(G, pos, ax=ax, with_labels=True, node_color='gray', edge_color='gray', node_size=6000, font_size=15, font_weight='bold')

# Set limits to ensure all nodes are within the viewable area
ax.set_xlim(-0.1, 0.3)
ax.set_ylim(-0.1, 0.3)

# Annotate nodes with dummy text
annotations = {
    'Member': 'Treatment',
    'CLV': 'Outcome of Interest',
    'Income': 'Potential Confounder'
}

for node, (x, y) in pos.items():
    ax.annotate(annotations[node], (x, y), textcoords="offset points", xytext=(0, 50), ha='center', fontsize=15, color='red')

# Save the figure
plt.savefig("/Users/ekinderdiyok/Documents/projects/causal-inference/diagram/causal_diagram.png", format="png", dpi = 300)
plt.savefig("/Users/ekinderdiyok/Documents/projects/causal-inference/diagram/causal_diagram.svg", format="svg")

# Reduce margins
plt.tight_layout()

plt.show()
```
